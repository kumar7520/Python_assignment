## *Binomial Distribution*

```python
## Binomial

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def binomial(n,p):
    x = []
    y = []
    for i in range(n+1):
        x.append(i)
        y.append(math.comb(n,i)*(p**i)*(1-p)**(n-i))
    print("mean = ", n*p)    
    #print("variance = ", np(1-p)) 
    plt.bar(x,y ,color =['red','green','yellow','blue','pink','orange','black'] )    
    plt.title(" binomial distribution")
    plt.xlabel("X")
    plt.show() 
binomial(10,0.5)
```

## *Normal Distribution*

```python
def plot_normal(u,s):
    x=np.linspace(u-4*s,u+4*s,1000)
    f=lambda x: (1/math.sqrt(2*math.pi*s**2))*np.exp(-0.5*((x-u)/s)**2)
    plt.plot(x,f(x))
    plt.title(f'Normal distribution with mean $\mu$={u} and variance $\sigma^2$={s**2}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_normal(3,2)
    
 
```

## *Poisson Distribution*

```python
def plot_poi(l):
    f=lambda x: np.exp(-l)*(l**x/math.factorial(x))
    x=[i for i in range(10)]
    y=[f(j) for j in x]
    plt.plot(x,y)
    plt.title(f'Poisson distribution with parameter $\lambda$={l}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_poi(10)
```
